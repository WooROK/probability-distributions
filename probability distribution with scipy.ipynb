### Binomial

```python
X = stats.binom(10,0.2) #Declare X to be a binomial random variable
print(X.pmf(3)) #P(X=3)
print(X.cdf(4)) #P(X<=4)
print(X.mean()) #E[X]
print(X.var()) #Var[X]
print(X.rvs()) # Get a random sample from X 
print(X.rvs(10)) #Get 10 random samples from X
#help(X) : List all methods defined for X 
```

```python
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10    
p = 0.2    

# 확률 변수 정의
X = range(n + 1)  # 0부터 n까지의 값
pmf = binom.pmf(X, n, p)

# 시각화
plt.stem(X, pmf, basefmt=" ")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial PMF (n={n}, p={p})")
plt.grid(True)
plt.show()
```

### Poisson
Make a Poisson Random variable Y
. This example declares Y∼Poi(λ=2)
. It then calculates P(Y=3)
:

```python
Y = stats.poisson(2)
print(Y.pmf(3)) #P(Y=3)
print(Y.rvs()) #Get a random sample from Y
```

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 파라미터 설정
mu = 3  # 평균(λ)

# 값 범위
x = range(0, 15)
pmf = poisson.pmf(x, mu)

# 시각화
plt.stem(x, pmf, basefmt=" ")  # 막대 그래프
plt.xlabel("Number of events (k)")
plt.ylabel("P(X = k)")
plt.title(f"Poisson PMF (λ = {mu})")
plt.grid(True)
plt.show()
```

### Normal 
A∼N(μ=3,σ^2=16)
In the Scipy library, the second parameter is the standard deviation, not the variance.

```python
A= stats.norm(3,math.sqrt(16)) #Declare A to be a normal randaom variable
print(A.pdf(4)) #f(4), the probability density at 4
print(A.cdf(4)) #F(4), which is also P(Y<4)
print(A.rvs()) #Get a random sample from A 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 3      # Mean
sigma = 4   # Standard deviation

# x values (continuous range)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Compute the PDF values
pdf = norm.pdf(x, mu, sigma)

# Plot
plt.plot(x, pdf, linewidth=2)
plt.title(f"Normal Distribution (μ={mu}, σ={sigma})")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

### Exponential 
B∼Exp(λ=4)


```python
# `λ` is a common parameterization for the exponential, 
# but `scipy` uses `scale` which is `1/λ`
B = stats.expon(scale = 1/4) 
print(B.pdf(2)) #the probability density at 2 
print(B.cdf(0.5)) #P(B<0.5)
print(B.rvs())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# λ = 4, so scale = 1/λ
rate = 4
scale = 1 / rate

# Create the distribution object
B = expon(scale=scale)

# x values
x = np.linspace(0, 3, 100)

# PDF values
pdf = B.pdf(x)

# Plot the PDF
plt.plot(x, pdf, 'b-', label=f'λ = {rate}')
plt.fill_between(x, pdf, alpha=0.3)
plt.title('Exponential Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```
